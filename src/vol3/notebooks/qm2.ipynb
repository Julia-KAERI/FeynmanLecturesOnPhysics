```julia
using CairoMakie
```

```julia
k_B = 1.380649e−23          # Boltzman constant for J/K
m_e = 9.109e-31             # electron mass in kg
m_p = 1.67262192e-27        # proton mass in kg
c = 299_792_458             # speed of light in m/sec
J2eV = 6.242e+18            # 1 J = 6.242e+18 eV
mu = 1.66039e-27            # atomic mass unit in kg
m_pi = m_e*273              # pi-meson mass
k_e = 8.987e9               # 1/4πϵ_0 in C^2⋅s^2⋅kg^{−1}⋅m^{−3}
q_e = 1.602176634e-19       # electron and proton charge in Coulombs.
ħ = 6.62607015e-34/2/π      # Planck constant in m^2 kg/s

G = 6.674e-11               # 만유인력 상수 in N⋅m^2⋅kg^{-2}
N_A = 6.02214076e23         # 아보가드로 수

pf2kg = 0.453592            # 1 Pf = 0.453592 kg
feet2m = 0.3048             # 1 feet = 0.3048 m
;
```

```julia
α = T -> ħ/k_B/T
ω = 1:1.0e10:1.0e14
```

```julia
ħ/k_B
```

```julia
d_T10 = @. exp(3*log(ω) - log(exp(α(10)*ω)-1));
d_T30 = @. exp(3*log(ω) - log(exp(α(30)*ω)-1));
```

```julia
f = Figure(backgroundcolor = RGBf(0.8, 0.8, 0.8),size = (1000, 700))
ax = Axis(f[1, 1], xlabel = "x label", ylabel = "y label",
    title = "Title")
lines!(ax, ω, d_T10 .* 1.0e3, label = "10 K")
lines!(ax, ω, d_T30 .* 10, label = "30 K")

axislegend(ax)
f
```
